```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["figure.dpi"] = 300
plt.style.use("seaborn")
```

```python
data = pd.read_csv("chap2_data.csv").x

data.head()
```

```python
len(data)
```

```python
data.describe()
```

```python
np.histogram(data, bins=7)
```

```python
plt.hist(data, bins=np.arange(-0.5, 8.5, 1))
```

```python
# sample variance
data.var(ddof=0)
```

```python
# sample standard deviation
data.std()
```

```python
y = range(10)
prob = stats.poisson.pmf(y, mu=3.56)

plt.plot(y, prob, "o--")
```

```python
fig, ax1 = plt.subplots()

ax1.hist(data, bins=np.arange(-0.5, 8.5, 1))
ax1.set_xlabel("data")
ax1.set_ylabel("Frequency")
ax2 = ax1.twinx()
ax2.plot(y, prob, "ro--")

plt.title("Histogram of data")
plt.show()
```
